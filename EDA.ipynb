```python
sns.boxplot(y='stress_level',x='employment_status',data=mental_health,palette='rocket')
plt.title('Stress Level By Employee Status')
plt.xlabel('Employee Status')
plt.ylabel('Strees Level')
plt.show()
```

```python
seek_treatment = mental_health['seeks_treatment'].value_counts()
print(seek_treatment)
```

```python
plt.pie(seek_treatment,autopct='%1.1f%%',startangle=90)
sns.set_theme()
plt.title('Propotion Seek Treatment')
plt.show()
```

```python
mental_health_risk = mental_health['mental_health_risk'].value_counts()
print(mental_health_risk)
```

```python
sns.barplot(x=mental_health_risk.index,y=mental_health_risk.values,palette='pastel')
plt.title('Distribution Of Mental Health')
plt.xlabel('Mental Health Risk')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
pd.crosstab(mental_health['gender'],mental_health['mental_health_risk']).plot(kind='barh', stacked=True, colormap='coolwarm')
plt.title('Mental Health Risk By Gender')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
```

```python
plt.scatter(mental_health['sleep_hours'],mental_health['stress_level'])
plt.title('Pattern Between Sleep And Stress')
plt.xlabel('Sleeping Hours')
plt.ylabel('Stress Level')
plt.show()
```

```python
pd.crosstab(mental_health['gender'],mental_health['employment_status']).plot(kind='barh')
plt.title('Gender Propotion With Employee Category')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.plot
```

```python
sns.boxplot(y='anxiety_score',x='gender',data=mental_health,palette='rocket')
plt.title('Stress Level By Employee Status')
plt.xlabel('Employee Status')
plt.ylabel('Strees Level')
plt.show()
```
